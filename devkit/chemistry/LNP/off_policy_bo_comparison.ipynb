```python
import torch
import numpy as np
from botorch.models import SingleTaskGP
from botorch.acquisition import UpperConfidenceBound, ProbabilityOfImprovement, ExpectedImprovement
from botorch.optim import optimize_acqf
import matplotlib.pyplot as plt

# Define your objective function
def objective_function(x):
    return -torch.sin(x*10) + -torch.sin(x*20)

# Set up initial data points
n_samples = 5
range_limit = 1
train_X = torch.rand(n_samples, 1).double() * range_limit  # Random initial points
train_Y = objective_function(train_X)
r = torch.arange(0,range_limit,0.001)
plt.plot(r,[objective_function(ri) for ri in r],color='k')
plt.scatter(train_X,train_Y,color='r')
plt.show()
```

```python
import torch
from gpytorch.constraints.constraints import GreaterThan
from gpytorch.kernels import MaternKernel, ScaleKernel
from gpytorch.likelihoods.gaussian_likelihood import GaussianLikelihood
from gpytorch.priors.torch_priors import GammaPrior
from botorch.fit import fit_gpytorch_mll
from botorch.models import SingleTaskGP
from botorch.models.transforms import Log
from botorch.optim import optimize_acqf
from botorch.test_functions import Ackley
from botorch.utils import standardize
from botorch.utils.sampling import draw_sobol_samples
from botorch.models.transforms.outcome import Standardize

from gpytorch.mlls import ExactMarginalLogLikelihood

MIN_INFERRED_NOISE_LEVEL = 1e-4


def get_matern_kernel_with_gamma_prior(
    ard_num_dims: int, batch_shape: None
) -> ScaleKernel:
    r"""Constructs the Scale-Matern kernel that is used by default by
    several models. This uses a Gamma(3.0, 6.0) prior for the lengthscale
    and a Gamma(2.0, 0.15) prior for the output scale.
    """
    return ScaleKernel(
        base_kernel=MaternKernel(
            nu=2.5,
            ard_num_dims=ard_num_dims,
            batch_shape=batch_shape,
            lengthscale_prior=GammaPrior(0.1, 0.6),
        ),
        batch_shape=batch_shape,
        outputscale_prior=GammaPrior(0, 0.15),
    )


#from sklearn.preprocessing import StandardScaler
class StandardScaler:

    def __init__(self, mean=None, std=None, epsilon=1e-7):
        """Standard Scaler.
        The class can be used to normalize PyTorch Tensors using native functions. The module does not expect the
        tensors to be of any specific shape; as long as the features are the last dimension in the tensor, the module
        will work fine.
        :param mean: The mean of the features. The property will be set after a call to fit.
        :param std: The standard deviation of the features. The property will be set after a call to fit.
        :param epsilon: Used to avoid a Division-By-Zero exception.
        """
        self.mean = mean
        self.std = std
        self.epsilon = epsilon

    def fit(self, values):
        dims = list(range(values.dim() - 1))
        self.mean = torch.mean(values, dim=dims)
        self.std = torch.std(values, dim=dims)

    def transform(self, values):
        return (values - self.mean) / (self.std + self.epsilon)

    def fit_transform(self, values):
        self.fit(values)
        return self.transform(values)

# Create a Gaussian Process model
    
gp = SingleTaskGP(train_X, train_Y,input_transform=Standardize) # covar_module=get_matern_kernel_with_gamma_prior(1,[1])

# Define the acquisition function (Upper Confidence Bound in this case)
UCB = UpperConfidenceBound(gp, beta=0.1)


# Optimization loop
n_iter = 30
acq_vals = []
for iteration in range(n_iter):  # Number of optimization steps

    # BEFORE sampling, what is the posterior
    mll = ExactMarginalLogLikelihood(likelihood=gp.likelihood, model=gp)
    fit_gpytorch_mll(mll)
    UCB = UpperConfidenceBound(gp, beta=0.1)

    X_preds = torch.linspace(0, 1, 100).unsqueeze(1)
    Y_preds = gp.posterior(X_preds)
    Y_mean = Y_preds.mean.squeeze().detach().numpy()
    Y_var = Y_preds.variance.squeeze().detach().numpy()



    # Optimize acquisition function to get the next point to evaluate
    candidate, acq_value = optimize_acqf(UCB, bounds=torch.tensor([[0.0], [range_limit]]), q=1, num_restarts=10, raw_samples=10)
    #print('--------------')
    #print(iteration)
    #print('candidate:',candidate)
    #print('Acquisition Value:',acq_value.numpy())
    acq_vals.append(acq_value.numpy())
    #print(UCB(candidate).detach())
    
    # Evaluate the objective function at the candidate point
    new_observation = objective_function(candidate)
    
    # Update the model with the new observation
    train_X = torch.cat([train_X, candidate])
    train_Y = torch.cat([train_Y, new_observation])

    ss = StandardScaler()
    ss.fit(train_Y)
    train_Y_scaled = ss.transform(train_Y)

    gp = SingleTaskGP(train_X, train_Y)#covar_module=get_matern_kernel_with_gamma_prior(1,[1])
    #UCB = UpperConfidenceBound(gp, beta=0.1)
    #UCB = ProbabilityOfImprovement(gp,best_f=1)
    
    # Get the best point found by the optimization loop
    #best_point = train_X[train_Y.argmax()]
    #print("Best Point:", best_point.item())

    plt.figure(figsize=(24,4))
    r = torch.arange(0,range_limit,0.001)

    plt.figure(figsize=(24,4))
    plt.plot(r,[objective_function(ri) for ri in r],color='k',linestyle='--',label='True Function')
    plt.plot(X_preds,Y_mean,color='m',linestyle='-',label='PosteriorMean')
    #plt.plot(X_preds,Y_mean + Y_var,color='r',linestyle='-',label='+Var')
    #plt.plot(X_preds,Y_mean - Y_var,color='g',linestyle='-',label='-Var')
    plt.fill_between(X_preds.numpy()[:, 0], Y_mean - Y_var, Y_mean + Y_var, color="m", alpha=0.5)
    plt.scatter(train_X[:n_samples],train_Y[:n_samples],color='k',marker='x',label='Initial Data')
    plt.xlabel("x")
    plt.ylabel("y")

    trials = []
    for index,(i,j) in enumerate(zip(train_X[n_samples:].numpy(),train_Y[n_samples:].numpy())):
        plt.text(i, j, f'{index}')
        trials.append(index)
    plt.scatter(train_X[n_samples:],train_Y[n_samples:],s=50,c=trials,zorder=2,label='BO Trials')
    plt.colorbar()
    plt.legend()

    plt.twinx()
    plt.plot(r,[UCB(torch.tensor([[ri]])).detach() for ri in r],color='r',linestyle='-',label='ACQ Function')
    plt.ylabel("Acquisition Value")
    plt.show()

    #plt.plot(trials,acq_vals)
    #plt.ylabel('Acquisition Value')
    #plt.xlabel('Trial')
    #plt.show()
```
