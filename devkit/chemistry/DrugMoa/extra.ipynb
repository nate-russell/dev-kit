```python
import tempfile
from devkit.dataset import fetch_file, unzip_to_data_subdir, get_subdir_files_as_dict
data = get_subdir_files_as_dict('MELIORA')
data
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(data['chemical-similarity.csv'])
df = df.set_index('broad_id')
df = df.sort_index()
df
```

```python
tdf = pd.read_csv(data["primary-screen-replicate-collapsed-treatment-info.csv"]).set_index('broad_id')
tdf = df.join(tdf)
tdf
```

```python
import seaborn as sns
XX = df.values
XX
```

```python
sns.histplot(XX.flatten())
plt.yscale('log')
plt.xlabel('Simililarity Distribution')
plt.show()

# Matrix
plt.imshow(XX)
plt.colorbar()
plt.show()

# Clusters
sns.clustermap(XX)
plt.show()
```

```python
import umap
reducer = umap.UMAP(
    verbose=True,
    n_neighbors=5,
    min_dist=0.25,
    spread=1.0,
    metric="precomputed",
    random_state=12,
    )
embedding = reducer.fit_transform(XX)
um1,um2 = embedding[:, 0], embedding[:, 1]
tdf['umap1'] = um1
tdf['umap2'] = um2
plt.scatter(tdf,x='um1',y='um2',hue='target',alpha=.7,marker='+')
```
