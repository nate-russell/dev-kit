# PIC50 vs IC50 w/ Plotting Examples
- interconversions
- making plots that don't require base 10 mental calculations

```python
from chem_conversions import pic50_to_ic50, ic50_to_pic50
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = 600
mode = 'nanomolar'
pic50 = ic50_to_pic50(x,'nanomolar')
print(f'{x} {mode} = {pic50} PIC50')

mode = 'micromolar'
ic50 = pic50_to_ic50(pic50,mode)
print(f'{pic50} PIC50 = {ic50:0.02f} {mode}')

x = 600
mode = 'micromolar'
pic50 = ic50_to_pic50(x,mode)
print(f'{x} {mode} = {pic50} PIC50')
```

## Plotting time

```python
a = ic50_to_pic50(1,'picomolar')
b = ic50_to_pic50(999,'milimolar')
px = np.linspace(a,b,25)
x = [pic50_to_ic50(v,'nanomolar') for v in px]
y = range(25)
```

```python
plt.figure()
fig1, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))

# nanomolar with log and chemistry friendly xtick labels
ax1.scatter(x, y)
ax1.set_xscale('log')
ax1.set_xticks([0.001,0.01,0.1,1,10,100,1000,10000,100000,1000000,10000000,100000000])
ax1.set_xticklabels(['1 pM','10 pM','100 pM','1 nM','10 nM','100 nM',u'1 \u03bcM',u'10 \u03bcM',u'100 \u03bcM','1 mM','10 mM','100 mM'],
                    rotation=90)
ax1.set_xlabel("$IC_{50}$")
ax1.set_ylabel("Index")
ax1.set_title("Example Plot with $IC_{50}$ X axis")

# nanomolar with log
ax2.scatter(x, y)
ax2.set_xlabel("$IC_{50}$ nM")
ax2.set_xscale('log')
ax2.set_ylabel("Index")
ax2.set_title("Example Plot with $IC_{50}$ nM X axis")

# pic50
ax3.scatter(px,y)
ax3.set_xlabel("$pIC_{50}$")
ax3.set_ylabel("Index")
ax3.set_title("Example Plot with $pIC_{50}$ X axis")
```
